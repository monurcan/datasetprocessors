```python
folder_name = "/mnt/RawInternalDatasets/KaraGAG/MWIR/FromWeb/ThermalSoccer/thermal-soccer-dataset/PublicDataset/images/4"
label_path = "/mnt/RawInternalDatasets/KaraGAG/MWIR/FromWeb/ThermalSoccer/thermal-soccer-dataset/PublicDataset/GT-4.xml"
output_path = "/mnt/RawInternalDatasets/KaraGAG/MWIR/FromWeb/ThermalSoccer/thermal-soccer-dataset/PublicDataset/labels/4"

IMAGE_WIDTH = 1920
IMAGE_HEIGHT = 480
```

```python
import os
from natsort import natsorted
from lxml import etree
import pathlib
```

```python
file_names = natsorted([file_.stem for file_ in list(pathlib.Path(folder_name).glob('*.png'))])
```

```python
with open(label_path, "rb") as xml_file:
    xml = xml_file.read()

root = etree.fromstring(xml)

for i, item in enumerate(root.getchildren()):
    if item.tag != "frame":
        continue
    
    print(item.attrib['num'])
    with open(os.path.join(output_path, f"{file_names[i]}.txt"), "w") as f:
        for record in item.getchildren():
            bb_coords = {key: float(coord) for key, coord in record.getchildren()[0].attrib.items()}
            #print(bb_coords)
            xMid = bb_coords['xb']
            yMid = bb_coords['yb']-bb_coords['h']/2
            
            yolo_bb = f"1 {xMid/IMAGE_WIDTH} {yMid/IMAGE_HEIGHT} {bb_coords['w']/IMAGE_WIDTH} {bb_coords['h']/IMAGE_HEIGHT}\n"

            f.write(yolo_bb)


```

```python
assert i==len(file_names)-1
```
